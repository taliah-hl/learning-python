```python
from matplotlib import pyplot as plt
fig = plt.figure(facecolor=(1, 0, 0, .1))   # red background to see where the figure is



ax = fig.add_subplot(111) # We'll explain the "111" later. Basically, 1 row and 1 column.
ax.set(xlim=[0.5, 4.5], ylim=[-2, 8], title='x graph')
plt.show()
az = fig.add_subplot(111)
az.set(xlim=[1,10],ylim = [-1,1], title='Z graph')
plt.show()
#cannot have 2 figure together!
```

```python
from matplotlib import pyplot as plt


fig, axes = plt.subplots(nrows=2, ncols=2)
'''呢句=
fig-plt.figure() <-create figure
ax = fig.add_subplot(22)
'''
axes[0,0].set(title='Upper Left')
axes[0,1].set(title='Upper Right')
axes[1,0].set(title='Lower Left')
axes[1,1].set(title='Lower Right')
# To iterate over all items in a multidimensional numpy array, use the `flat` attribute
for ax in axes.flat:
    # Remove all xticks and yticks...
    ax.set(xticks=[], yticks=[])
#print(axes)
print(fig)
plt.show()
```

同樣道理,假設你想plot cos x, sin x, tan x 三個圖
有2個方法
1. create 1 figure, 3 subplots
2. create 3 plots
下面先示範用3個subplot畫

```python
import numpy as np
import matplotlib.pyplot as plt

# Try to reproduce the figure shown in images/exercise_1-1.png

# Our data...
x = np.linspace(0, 10, 100)
y1, y2, y3 = np.cos(x), np.sin(x), np.tan(x)
names = ['Signal 1', 'Signal 2', 'Signal 3']

fig, axes = plt.subplots(nrows = 3, ncols = 1)
axes[0].set(title='Cos x')
axes[1].set(title='sin x')
axes[2].set(title='tan x')
axes[0].plot(x, y1)
axes[1].plot(x, y2)
axes[2].plot(x, y3)

for ax in axes.flat:
    ax.set(xticks=[])
plt.show()
```
